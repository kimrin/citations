```python
import pandas as pd
# Set display options to show more rows
pd.set_option('display.max_rows', None)  # None means no limit
pd.set_option('display.max_columns', None)  # Show all columns if needed

df = pd.read_csv('female_seiyuu_list.csv')
df
```

```python
df2 = df.sort_values(by='出典数', ascending=False)
df2['出典数順位'] = df2['出典数'].rank(method='min', ascending=False).astype(int)
df2.reset_index(inplace=True)
df3 = df2[['順位', '出典数順位', '声優名', '出典数']]
df3
```

```python

df3['advanced'] = df3['順位'] - df3['出典数順位']
df3['fan_level'] = df3['advanced'].rank(method='min', ascending=False).astype(int)
df4 = df3.sort_values(by='fan_level', ascending=True)

df4.to_csv('female_seiyuu_list_with_fan_level.csv', index=False)
```
